```python
import requests
import json
import pandas as pd
import unidecode
```

```python
exemplo = 'https://www.camara.leg.br/propostas-legislativas/2297446'
numero = exemplo.split("-legislativas/",1)[1]
numero
```

```python
prefixo_pl = 'https://dadosabertos.camara.leg.br/api/v2/proposicoes/'
sufixo_pl = '/autores'
```

```python
url = prefixo_pl+numero+sufixo_pl
url
```

```python
dados_gerais = []
parametros = {'formato': 'json'}
resposta = requests.get(url, parametros)
for lista in resposta.json()['dados']:
    dicionario = {"uri": lista['uri'],
                  "nome": lista['nome'],
                 }
    dados_gerais.append(dicionario)
df_teste = pd.DataFrame(dados_gerais)

lista1 = []
lista2 = []
for num, row in df_teste.iterrows():
    endereco = row['uri']
    cod = endereco.split("/deputados/",1)[1]
    cod = cod+"|"
    
    nome = row['nome']+"|"
    
    lista1.append(cod)
    lista2.append(nome)
```

```python
str1 = ''.join(lista2)
str1 = str1[:-1]
```

```python
proposicoes = pd.read_excel('resultados/proposicoes/Planilha de projetos - Ruralômetro - versão revisada.xlsx', sheet_name='Sheet1', dtype='str')
```

```python
proposicoes.rename(columns = {'Número do projeto':'nome_pl'},inplace = True)
proposicoes.rename(columns = {'Filtros':'filtros'},inplace = True)
proposicoes.rename(columns = {'Resumo':'resumo'},inplace = True)
proposicoes.rename(columns = {'Entidade avaliadora':'entidade_avaliadora'},inplace = True)
proposicoes.rename(columns = {'Avaliação':'avaliacao'},inplace = True)
proposicoes.rename(columns = {'Autor(es)':'autor_es'},inplace = True)
```

```python
proposicoes
```

```python
pls_finais = []

for num, row in proposicoes.iterrows():
    endereco = row['site_proposicao']
    print(endereco)
    
    try:
        numero = endereco.split("-legislativas/",1)[1]
    except:
        numero = endereco.split("idProposicao=",1)[1]
        
    
    nome_pl	= row['nome_pl']
    filtros = row['filtros']
    ementa = row['ementa']
    resumo = row['resumo']
    entidade_avaliadora	= row['entidade_avaliadora']
    avaliacao = row['avaliacao']
    urlInteiroTeor = row['urlInteiroTeor']
    site_proposicao = row['site_proposicao']
    
    
    url = prefixo_pl+numero+sufixo_pl
    #print(url)
    
    # Captura ids e autores da proposcicao e coloca em dataframe
    dados_gerais = []
    parametros = {'formato': 'json'}
    resposta = requests.get(url, parametros)
    for lista in resposta.json()['dados']:
        dicionario = {"uri": lista['uri'],
                  "nome": lista['nome'],
                 }
        dados_gerais.append(dicionario)
    df_teste = pd.DataFrame(dados_gerais)
    
    #Separa conteudo em uma string
    lista1 = []
    lista2 = []
    for num, row in df_teste.iterrows():
        local_deput = row['uri']
        
        try:
            cod = local_deput.split("/deputados/",1)[1]
        except:
            cod = local_deput.split("/orgaos/",1)[1]
            
        cod = cod+"|"
    
        nome = row['nome']+"|"
    
        lista1.append(cod)
        lista2.append(nome)
    str1 = ''.join(lista1)
    str1 = str1[:-1]
    str2 = ''.join(lista2)
    str2 = str2[:-1]
    
    #Cria dataframe final
    dicionario = {"nome_pl": nome_pl,
                  "filtros": filtros,
                  "ementa": filtros,
                  "resumo": resumo,
                  "entidade_avaliadora": entidade_avaliadora,
                  "avaliacao": avaliacao,
                  "urlInteiroTeor": urlInteiroTeor,
                  "site_proposicao": site_proposicao,
                  "autor_es": str2,
                  "id_interno_autores": str1                                
                  
                 }
    pls_finais.append(dicionario)

df_pls_finais = pd.DataFrame(pls_finais)
```
