```python
import pandas as pd
```

```python
bancada = pd.read_excel('originais/bancada_ruralista_8_mar.xlsx',sheet_name='Planilha1')
```

```python
bancada.sample(5)
```

```python
bancada['nome_deputado'] = bancada['nome_deputado'].str.strip()
bancada['nome_deputado'] = [row.replace(' *', '') for row in bancada['nome_deputado']]
```

```python
bancada.sample(5)
```

```python
kwargs = {'sep': ',', 'dtype': str, 'encoding': 'utf-8'}
politicos = pd.read_csv("resultados/politicos_2018_09abr.csv", **kwargs)
```

```python
politicos
```

```python
politicos_bancada = pd.merge(politicos, bancada, left_on='nome_urna', right_on='nome_deputado', how = 'right')
```

```python
politicos_bancada
```

```python
politicos_bancada.to_excel('resultados/checa_bancada.xlsx',sheet_name='Sheet1',index=False)
```

```python
politicos_bancada_new = pd.read_excel('resultados/checa_bancada.xlsx', dtype = str, sheet_name='Sheet1')
```

```python
politicos_bancada_new
```

```python
politicos_bancada_new = politicos_bancada_new[['bancada_ruralista', 'politico_id']]
```

```python
politicos_bancada_final = pd.merge(politicos, politicos_bancada_new, left_on='politico_id', right_on='politico_id', how = 'left')
```

```python
politicos_bancada_final
```

```python
politicos_bancada_final.to_excel('resultados/politicos_2018_09abr.xlsx',sheet_name='Sheet1',index=False)
```

```python
politicos_bancada_final.to_csv('resultados/politicos_2018_09abr.csv',index=False)
```
